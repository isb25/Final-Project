Next, we will look at some datasets to determine which neighborhood has the most options for Outdoor Recreation. Options that were considered were number of playgrounds, number of courts/rinks(basketball, tennis, hockey, etc.), number of playing fields(baseball, soccer, football), and number of pools. 

```python
import pandas as pd
from pandas import DataFrame
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
import itertools  
total_dict = {}
```

```python
playgrounds = pd.read_csv('https://data.wprdc.org/datastore/dump/47350364-44a8-4d15-b6e0-5f79ddff9367')
part1 = playgrounds['neighborhood'].value_counts().to_dict()
plist = playgrounds['neighborhood'].value_counts().head(13).to_dict()
df = DataFrame(list(plist.items()), columns = ['Number of Playgrounds','Neighborhood']) 
dgraph = df.plot.bar(x='Number of Playgrounds', y='Neighborhood')
```

This graph shows the neighborhoods with the most amount of playgrounds

```python
courts = pd.read_csv('https://data.wprdc.org/datastore/dump/a5b71bfa-840c-4c86-8f43-07a9ae854227')
part2 = courts['neighborhood'].value_counts().to_dict()
clist = courts['neighborhood'].value_counts().head(14).to_dict()
cf = DataFrame(list(clist.items()), columns = ['Number of courts','Neighborhood'])
cgraph = cf.plot.bar(x='Number of courts', y='Neighborhood')
```

This graph shows the neighborhoods with the most amount of courts/rinks.

```python
fields = pd.read_csv('https://data.wprdc.org/datastore/dump/6af89346-b971-41d5-af09-49cfdb4dfe23')
part3 = fields['neighborhood'].value_counts().to_dict()
flist = fields['neighborhood'].value_counts().head(15).to_dict()
ff = DataFrame(list(flist.items()), columns = ['Number of fields','Neighborhood'])
fgraph = ff.plot.bar(x='Number of fields', y='Neighborhood')
```

This graph shows the neighborhoods with the most playing fields

```python
pools = pd.read_csv('https://data.wprdc.org/datastore/dump/5cc254fe-2cbd-4912-9f44-2f95f0beea9a')
part4 = pools['neighborhood'].value_counts().to_dict()
poolist = pools['neighborhood'].value_counts().to_dict()
poolf = DataFrame(list(poolist.items()), columns = ['Number of pools','Neighborhood'])
poolgraph = poolf.plot.bar(x='Number of pools', y='Neighborhood')
```

This graph shows all of the neighborhoods with pools

```python
total_dict = part1
for key in part2:
    if key in total_dict:
        total_dict[key] = total_dict[key] + part2[key]
    else:
        total_dict[key] = part2[key]

for key in part3:
    if key in total_dict:
        total_dict[key] = total_dict[key] + part3[key]
    else:
        total_dict[key] = part3[key]

for key in part4:
    if key in total_dict:
        total_dict[key] = total_dict[key] + part4[key]
    else:
        total_dict[key] = part4[key]


total_dict = {k: v for k, v in sorted(total_dict.items(), key=lambda item: item[1], reverse = True)}

```

```python
final = dict(itertools.islice(total_dict.items(), 10))  
```

```python
print(final)
```

```python
finalframe = DataFrame(list(final.items()), columns = ['Most Outdoor Recreation','Neighborhood'])
finalgraph = finalframe.plot.bar(x='Most Outdoor Recreation', y='Neighborhood')
```
