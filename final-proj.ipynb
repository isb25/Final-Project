```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
neighborhood = pd.read_csv("Neighborhoods_.csv")
hoodsdict = neighborhood['hood'].value_counts().to_dict()
neighborhood.set_index('hood')

incidents = pd.read_csv("Police Incident .csv")
incidentdict = incidents['INCIDENTNEIGHBORHOOD'].value_counts().to_dict()



score = hoodsdict

for key in incidentdict:
    if key in score:
        score[key] = score[key] + incidentdict[key]
        
array = list(score.values())
array.sort(reverse = True)
ratio = 50/array[0]
for key in score:
    score[key] = ratio * score[key]
s = pd.Series(score, name="Score")
s.index.name = "Neighborhood"
plt.ylabel("Number of incidents")
s.sort_values(ascending=False).iloc[0:12].plot(kind='bar')

```

```python
neighborhood = pd.read_csv("Neighborhoods_.csv")
hoodsdict = neighborhood['hood'].value_counts().to_dict()
neighborhood.set_index('hood')

arrests = pd.read_csv("Arrests.csv")
arrestsdict = arrests['INCIDENTNEIGHBORHOOD'].value_counts().to_dict()

scoresArrests = hoodsdict

for key in arrestsdict:
    if key in scoresArrests:
        scoresArrests[key] = scoresArrests[key] + arrestsdict[key]
array = list(scoresArrests.values())
array.sort(reverse=True)
ratio = 50/array[0]

for key in scoresArrests:
    scoresArrests[key] = ratio * scoresArrests[key]
s = pd.Series(scoresArrests, name="Score")
s.index.name = "Neighborhood"
plt.ylabel("Number of Arrests")
s.sort_values(ascending=False).iloc[0:12].plot(kind='bar')
```
